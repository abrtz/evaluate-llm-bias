```python
import pandas as pd
import csv
import os
import collections
```

```python
file_path = "/content/drive/MyDrive/Colab Notebooks/historico-nombres.csv"
```

```python
#initialize an empty dictionary to store name totals
name_totals = {}

with open(file_path, mode='r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        #strip leading/trailing spaces from 'nombre'
        name = row['nombre'].strip().title()
          #only interested in single word names
        if len(name.split()) == 1:
          #update name_totals
          name_totals[name] = name_totals.get(name, 0) + int(row['cantidad'])
```

```python
#order the dictionary by values (cantidad) in descending order
ordered_name_totals = collections.OrderedDict(sorted(name_totals.items(), key=lambda item: item[1], reverse=True))
```

```python
gendered_names = []

#counters for each gender
male_count = 0
female_count = 0

for name, total in ordered_name_totals.items():
    name_dict = {}
    if name.endswith('a'):
        gender = 'female'
    else:
        gender = 'male'

    #limit to first 15 instances per gender
    if gender == 'female' and female_count < 15:
        name_dict['name'] = name
        name_dict['gender'] = gender
        gendered_names.append(name_dict)
        female_count += 1

    elif gender == 'male' and male_count < 15:
        name_dict['name'] = name
        name_dict['gender'] = gender
        gendered_names.append(name_dict)
        male_count += 1
```

```python
file_path = "/content/drive/MyDrive/Colab Notebooks/top_names.csv"

os.makedirs(os.path.dirname(file_path), exist_ok=True)

#write the names to a csv file
with open(file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['name', 'gender'])
    for item in gendered_names:
        writer.writerow([item['name'], item['gender']])

print("Names have been written to 'top_names.csv'")
```
