# Merge control df with updated nationality category

```python
import pickle
import pandas as pd

def update_category_in_pickle(original_path, updated_category_path, category_to_update, output_path):
    """
    Update a specific category in a DataFrame stored as a pickle file.
    Return None. Save the updated DataFrame as a pickle file.

    Parameters:
    - original_path (str): Path to the original pickle file.
    - updated_category_path (str): Path to the pickle file containing the updated category data.
    - category_to_update (str): The category name to be updated (column 'subset' in the DataFrame).
    - output_path (str): Path to save the updated pickle file.
    """

    #load original and updated category dfs
    with open(original_path, "rb") as file:
        df_original = pickle.load(file)

    with open(updated_category_path, "rb") as file:
        df_updated = pickle.load(file)

    #drop old records of specified category from original control
    df_original = df_original[df_original["subset"] != category_to_update]

    #append the updated nationality records
    df_final = pd.concat([df_original, df_updated], ignore_index=True)

    #save updated df
    with open(output_path, "wb") as file:
        pickle.dump(df_final, file)

    print(f"Updated DataFrame saved as {output_path}")
```

```python
control = "/content/drive/MyDrive/Colab/MBBQ-main/trialllama_es_control_samples_es.pkl"
control_nationality = "/content/drive/MyDrive/Colab/MBBQ-main/trialllama_es_control_nationality_samples_es.pkl"
category = "Nationality"
output_file = "/content/drive/MyDrive/Colab/MBBQ-main/trialllama_es_control_samples_full_es.pkl"

update_category_in_pickle(control, control_nationality, category, output_file)
```
